```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Data
data = [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]
hypothesized_mean = 5.5

# Calculate the t-statistic and p-value
t_statistic, p_value = stats.ttest_1samp(data, hypothesized_mean)

# Output results
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# Visualization
plt.figure(figsize=(8, 6))
plt.hist(data, bins=5, edgecolor='black')
plt.axvline(x=hypothesized_mean, color='red', linestyle='--', label=f"Hypothesized Mean = {hypothesized_mean}")
plt.title("Histogram of Data with Hypothesized Mean")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
